## more complex generator

generator always yield, stop and resume where it is left off.

```python
def gen456():
    print('about to yield 2')
    yield 2
    print('about to yield 4')
    yield 4
    print('about to yield 6')
    yield 6
    print('about to return')
```

```python
g = gen456()
```

```python
next(g)
```

```python
next(g)
```

```python
next(g)
```

```python
next(g)
```

## 3) Laziness and Infinite

```python
# lucas : 2,1,3,4,....
# first two values are 2,1
# then sum of preceeding two numbers

def lucas():
    yield 2
    a = 2
    b = 1
    while True:
        yield b
        a,b = b,a+b
```

```python
lucas()
```

```python
for item in lucas():
    print(item)
```

## Generator Expressions

(expr(item) for item in iterable)

```python
million_squares = (x*x for x in range(1,1000001))
million_squares
```

```python
list(million_squares)[-10:]
```

```python
#now generator object is exhausted once called
list(million_squares)
```
